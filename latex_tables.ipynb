```python
FRAMES_STRUCTURES = {
    "111,110": [("X", "Z", "Z"), ("Z", "X", "X")],
    "111,011": [("Z", "Z", "X"), ("X", "X", "Z")],
    "110,110": [("X", "Z", "X"), ("Z", "X", "Z")],
    "110,011": [("X", "Z", "X"), ("Z", "X", "Z")],
    "101,111": [("Z", "Z", "X"), ("X", "Z", "Z")],
    "101,010": [("Z", "X", "X"), ("X", "X", "Z")],
    "101,100": [("X", "X", "Z"), ("X", "Z", "Z")],
    "101,001": [("Z", "X", "X"), ("Z", "Z", "X")],
    "100,111": [("X", "X", "X"), ("Z", "X", "Z")],
    "100,100": [("X", "X", "X"), ("X", "Z", "X")],
    "001,011": [("Z", "X", "X"), ("X", "Z", "Z")],
    "100,010": [("X", "X", "X"), ("Z", "X", "Z")],
    "100,001": [("X", "X", "X"), ("X", "Z", "X")],
    "011,111": [("X", "X", "Z"), ("Z", "X", "X")],
    "011,100": [("Z", "X", "X"), ("Z", "Z", "X")],
    "011,010": [("X", "Z", "Z"), ("Z", "Z", "X")],
    "011,001": [("X", "Z", "Z"), ("X", "X", "Z")],
    "010,111": [("X", "Z", "X"), ("Z", "Z", "Z")],
    "010,100": [("Z", "X", "Z"), ("Z", "Z", "Z")],
    "010,010": [("Z", "Z", "Z"), ("X", "Z", "X")],
    "010,001": [("Z", "Z", "Z"), ("Z", "X", "Z")],
    "001,110": [("Z", "Z", "X"), ("X", "X", "Z")],
    "000,110": [("Z", "Z", "Z"), ("X", "X", "X")],
    "000,011": [("Z", "Z", "Z"), ("X", "X", "X")]
}

def get_data(s):
    s_data = []
    for struct in FRAMES_STRUCTURES[s]:
        temp = []
        for bit, base in zip(s[4:], struct):
            if base == "Z":
                temp.append(f"- & {bit}_{base}")
            elif base == "X":
                temp.append(f"{bit}_{base} & -")
        s_data.append(temp)
    return s_data

def generate_latex_code(results):
    if len(results) == 6:
        x, y = 1, 1
        v, w = 1, 2

    else:
        if results[-1] == "IDLE_DOWN-MOVE":
            x, y = 2, 1
            v, w = 2, 2
        elif results[-1] == "MOVE-IDLE":
            x, y = 1, 1
            v, w = 2, 1

    s_j_data = get_data(results[0])
    s_k_data = get_data(results[1])
    
    results_data = ["", ""]
    if results[2] and results[4]:
        results_data[0] = "$f^{s_" + str(x) + "}_{j} = f^{s_" + str(y) + "}_{k} \implies MR_j = " + MATCHING_RESULTS[results[2]] + ", MR_k = " + MATCHING_RESULTS[results[4]] + "$"
    
    if results[3] and results[5]:
        results_data[1] = "$f^{s_" + str(v) + "}_{j} = f^{s_" + str(w) + "}_{k} \implies MR_j = " + MATCHING_RESULTS[results[3]] + ", MR_k = " + MATCHING_RESULTS[results[5]] + "$"
        
    latex_code = f"""
    \\hline & \\\\[-2.0ex]
    % {results[0]} - {results[1]}
    \\multirow{{4}}{{*}}{{$\\underset{{S_j = {results[0]}}}{{
        \\begin{{pmatrix}}
            {s_j_data[0][0]} \\\\
            {s_j_data[0][1]} \\\\
            {s_j_data[0][2]}
        \\end{{pmatrix}} , 
        \\begin{{pmatrix}}
            {s_j_data[1][0]} \\\\
            {s_j_data[1][1]} \\\\
            {s_j_data[1][2]}
        \\end{{pmatrix}}
    }}
    \\text{{ }}
    -
    \\text{{ }}
    \\underset{{S_k = {results[1]}}}{{
        \\begin{{pmatrix}}
            {s_k_data[0][0]} \\\\
            {s_k_data[0][1]} \\\\
            {s_k_data[0][2]}
        \\end{{pmatrix}} , 
        \\begin{{pmatrix}}
            {s_k_data[1][0]} \\\\
            {s_k_data[1][1]} \\\\
            {s_k_data[1][2]}
        \\end{{pmatrix}}
    }}$}} & {results_data[0]} \\\\
    &  \\\\
    & {results_data[1]} \\\\
    & \\\\
            """
    return latex_code

for d in data:
    print(generate_latex_code(d))
```
