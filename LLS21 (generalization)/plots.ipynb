```python
depolarize_probability = 1

df = pd.read_csv(f"results_{int(depolarize_probability * 100)}%.csv")[:1000]
df.plot.scatter(x = "double_matchings", y = "%", xlabel = "double matchings", ylabel = "bits recovered ratio", alpha = 0.5, grid = True, marker = ".", s = 50)

plt.savefig('figures/LLS21_generalization_results.svg', format = "svg")
```

```python
df.plot.scatter(x = "double_matchings", y = "time", xlabel = "double matchings", ylabel = "execution time (seconds)", alpha = 0.5, grid = True, marker = ".", color = "green", s = 50)

plt.savefig('figures/LLS21_generalization_time_results.svg', format = "svg")
```
